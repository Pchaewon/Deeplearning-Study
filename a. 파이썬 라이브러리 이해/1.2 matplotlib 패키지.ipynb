##  matplotlib : 다양한 데이터의 시각화 기능을 제공
> pyplot : 서브패키지. 수치해석 소프트웨어로 알려진 matlab이라는 시각화 명령을 거의 그대로 사용할 수 있도록 하는 API 명령어 집합을 제공 </br>
> matplotlib의 별칭 : mpl </br>
> matplotlib.pyplot의 별칭 : plt </br>
> 주피터 노트북은 show() 명령어 없이도 셀 단위 자동 렌더링 됨. 인터프리터는 항상 show()를 마지막에 추가해야됨.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4,2)) #그림의 x축과 y축의 크기 비율
x = [10, 20, 30, 40]
y = [1, 4, 9, 16]
plt.plot(x, y, "b--")
plt.plot(x, y,"r^")
plt.grid(True, ls="--", lw=1)
#plt.show()
```

```python
plt.figure(figsize=(4,2))
plt.plot([10,20,30,40],[1,4,9,16],c="b", lw=2, ls=":", ms=15, marker="o", mew=5, mec="g", mfc="r")
plt.grid(True, ls="--", lw=1)
plt.show()
```

```python
import matplotlib as mpl
import numpy as np

mpl.rc('font', family='Nanumgothic')
mpl.rc('axes', unicode_minus=False)

plt.figure(figsize=(4,2))
x=np.arange(-3.14, 3.14, 0.1)

plt.title("sin() & cos() 함수", size =14)
plt.xlabel("각도(Angle)", size=12)
plt.ylabel("sin(x)",size=12)
plt.grid(True, ls=":", lw=1)
plt.plot(x, np.sin(x), 'b-', label="sin()")
plt.plot(x, np.cos(x), 'r--', label="cos()")
plt.legend(loc="upper right")
plt.show()
```
